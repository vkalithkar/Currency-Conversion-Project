```python
import pandas as pd
import os
```

```python
file_path = os.path.join('..', 'data', 'exchange_data.csv')

exchange_df = pd.read_csv(file_path)
exchange_df
```

```python
exchange_df[['exchange_rate','currency_name']] = exchange_df['currency'].str.split(' ',n=1, expand=True)
exchange_df.drop('currency', axis = 1, inplace = True)
```

```python
exchange_df
```

```python
exchange_df['country'] = exchange_df['country'].str.split(',',n=1, expand=True)[0]
exchange_df
```

```python
exchange_df = exchange_df.dropna()
```

```python
exchange_df['year'] = exchange_df['year'].astype(int)
```

```python
exchange_df
```

```python
file_path_new = os.path.join('..', 'data', 'clean_exchange_data.csv')
exchange_df.to_csv(file_path_new)
```

```python
import pandas as pd
import os
```

```python
file_path = os.path.join('..', 'data', 'clean_exchange_data.csv')

exchange_df = pd.read_csv(file_path)
exchange_df
```

```python
exchange_df = exchange_df.apply(lambda x: x.str.strip() if x.dtype == "object" else x)
```

```python
exchange_df.to_csv(os.path.join('..', 'data', 'cleaned_exchange_data.csv'), index = False)
```

```python
print(exchange_df.loc[exchange_df['country'] == 'Denmark', 'currency_name'].unique())
```

```python
exchange_df.loc[exchange_df['country'] == 'Denmark', 'currency_name'] = \
    exchange_df.loc[exchange_df['country'] == 'Denmark', 'currency_name'].str.lower()
```

```python
exchange_df.loc[exchange_df['country'] == 'Denmark', 'currency_name'] = \
    exchange_df.loc[exchange_df['country'] == 'Denmark', 'currency_name'].str.capitalize()
```

```python
print(exchange_df.loc[exchange_df['country'] == 'Denmark', 'currency_name'].unique())
```

```python
exchange_df.to_csv(os.path.join('..', 'data', 'clean_exchange_data.csv'), index = False)
```

```python
print(exchange_df.loc[exchange_df['country'] == 'Norway', 'currency_name'].unique())
```

```python
exchange_df.loc[exchange_df['country'] == 'Norway', 'currency_name'] = 'Krone'
```

```python
exchange_df.to_csv(os.path.join('..', 'data', 'clean_exchange_data.csv'), index = False)
```
