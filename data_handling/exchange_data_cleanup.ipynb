```python
import pandas as pd
import os

# Assuming exchange_df already exists (your original dataset)
file_path = os.path.join('..', 'data', 'clean_exchange_data.csv')

exchange_df = pd.read_csv(file_path)

# Create a new DataFrame for US data
years = list(range(1813, 2011))  # 1813 to 2010 inclusive

usa_data = pd.DataFrame({
    "country": ["United States of America"] * len(years),
    "year": years,
    "exchange_rate": [1.0] * len(years),
    "currency_name": ["United States Dollar"] * len(years)
})

# Append the new USA data to your existing exchange_df
exchange_df = pd.concat([exchange_df, usa_data], ignore_index=True)

# (Optional) Sort if you want neatness
exchange_df = exchange_df.sort_values(by=["country", "year"]).reset_index(drop=True)
```

```python
import pandas as pd
import os

# Assuming exchange_df already exists (your original dataset)
file_path = os.path.join('..', 'data', 'clean_exchange_data.csv')

exchange_df = pd.read_csv(file_path)
```

```python
# Group by country and currency_name, get min/max year
currency_ranges = (
    exchange_df
    .groupby(['country', 'currency_name'])['year']
    .agg(['min', 'max'])
    .reset_index()
)

# Build the formatted range string
def format_label(row):
    if row['max'] >= 2025:
        return f"{row['currency_name']} ({row['min']}-present)"
    elif row['min'] == row['max']:
        return f"{row['currency_name']} ({row['min']})"
    else:
        return f"{row['currency_name']} ({row['min']}-{row['max']})"

currency_ranges['currency_range'] = currency_ranges.apply(format_label, axis=1)

# Merge back
exchange_df = exchange_df.merge(
    currency_ranges[['country', 'currency_name', 'currency_range']],
    on=['country', 'currency_name'],
    how='left'
)
```
